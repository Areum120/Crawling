### 자동으로 검색하기 (팝업창 끄기) ###

```python
#Step 0. 필요한 모듈과 라이브러리를 로딩하고 검색어를 입력 받습니다
from bs4 import BeautifulSoup
from selenium import webdriver
import time#time.sleep 명령어 실행하기 위해 import
```

```python
# query_txt = input('크롤링할 키워드는 무엇입니까?:')
# 사용자에게 검색어 입력

query_txt = '코로나'
```

```python
# - path에는 chromedriver.exe 파일의 경로를 적어준다.
# - path의 경로에서 chromedriver를 실행하여 크롬 프로그램으로 관광공사의
#   웹 페이지를 연다
# - driver.get("URL 주소")
# - time.sleep(2)는 웹 페이지가 모두 열릴 때 까지 2초 기다린다.
#Step 1. 크롬 드라이버를 사용해서 웹 브라우저를 실행합니다.
path = "C:\study\python\datadown\chromedriver.exe"
driver = webdriver.Chrome(path)
#크롬으로 웹페이지를 열기
```

```python
driver.get("https://www.data.go.kr/")
time.sleep(2)

#위 페이지가 모두 열릴 때까지 2초 대기
```

```python
#driver.maximize_window()
# 코로나 얼럿창 있을 경우 닫기 클릭하기 //*[@id="chkForm01"]는 xpath의 값

try:
    driver.find_element_by_xpath('//*[@id="layer_popup_info_1"]/div[1]/a').click()
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="layer_popup_info_0"]/div[1]/a').click()
except:
    print("코로나 창이 없습니다.")
```

```python
#Step 2. 검색창의 이름을 찾아서 검색어를 입력합니다
element = driver.find_element_by_id("keyword")
element.send_keys(query_txt)#검색어입력하게 하는 부분
element.send_keys("\n")#엔터효과
# ※ XPath(XML Path Language)
# - XPath는 XML 문서의 특정 부분의 위치를 찾을 때 사용하는 언어이다.
# - XPath는 XML 문서안에 요소와 속성을 탐색하다.
# - '/'는 문서의 계층 구조에서 최상위 노드 역할을 하는 가상 노드이다.
# - XPath에서 속성을 지정하려면 @를 사용한다
# - //* 는 문서 내의 모슨 요소를 선택한다. 
# - [@id="gnb"] 는 id='gnb' 인것을 선택한다.
# - /div/div[1] 는 /div 자식인 첫번째 div요소

# ※ 개발자 도구에서 XPath 가져오기
# - 개발자 도구 F12 -> 1번 클릭 -> 파업창의 닫기 버튼 클릭(2번) -> 3번은 마우스 오른쪽 클릭
# - 펼쳐진 메뉴에서 Copy 선택 -> Copy XPath 선택하면 XPath값이 클립모드에 복사된다.
# - driver.find_element_by_xpath() 함수 인자값에 붙여넣기 한다.
```

### 검색 결과에서 텍스트 데이터만 추출하기 ###

```python
# Step 4. 현재 페이지에 있는 내용을 화면에 출력하기
 
time.sleep(1)
 
full_html = driver.page_source
# 검색결과 페이지의 전체 HTML 소스코드 가져온다
 
soup = BeautifulSoup(full_html, 'html.parser')
# 전체 HTML 소스코드를 Beautiful Soup이 분석하도록 지정하는 코드


# - 개발자도구(F12)에서 검색된 결과 각각이 <ul class='list_thumType type1'> 코드 아래
#   여러개의 li가 있으므로 Beautiful Soup에게 <ul class='list_thumType type1'>아래의 내용만 다 가져오게
#   한후 각각의 li 태그 안에 있는 텍스트들을 반복문을 사용하여 추출한다.
 
# content_list = soup.find('ul','list_thumType type1')


content_list = soup.find_all('div','result-list') 

 
for i in content_list:
    print(i.text.strip())
    print("\n")
```

### 2. 추출한 텍스트 저장하기 ###

```python


# - 위에서 추출한 텍스트 데이터를 파일에 저장한다.
# - 화면에 출력된 결과를 화면에 보이지 않고 파일로 저장한다.
# - 표준 출력 방향은 모니터인데 그 표준 출력 방향을 파일로 바꾸어서 
#   파일에 출력(저장)한다.

# - orig_stdout = sys.stdout 는 표준 출력방향을 다른변수로 지정한다.
# - 아래 코드는 표준 출력 방향에 이어쓰기 모드의 파일객체로 지정한다.
#   f = open(f_name , 'a' , encoding='UTF-8')
#   sys.stdout = f 

# - sys.stdout의 print함수는 이제 모니터에 출력하지 않고 파일에 저장한다.

# 학습목표 2: 텍스트를 추출하여 txt 형식으로 저장하기
# Step 5. 현재 페이지에 있는 내용을 txt 형식으로 파일에 저장하기

f_name = input('검색 결과를 저장할 파일경로와 이름을 지정하세요(예:C:\\study\\python\\notebook\data\\test.txt): ')

import sys
orig_stdout = sys.stdout 
f = open(f_name , 'a' , encoding='UTF-8')
sys.stdout = f
time.sleep(1)
 
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
content_list = soup.find_all('div','result-list') 

 
for i in content_list:
    print(i.text.strip())
    print("\n")
    
sys.stdout = orig_stdout
f.close()
 
print(" 요청하신 데이터 수집 작업이 정상적으로 완료되었습니다")
```
